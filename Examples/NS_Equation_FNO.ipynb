```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from neuralop import get_model, LpLoss, H1Loss
from neuralop.training import Trainer, AdamW
from neuralop.utils import count_model_params
from neuralop.data.datasets import load_navier_stokes_pt
```

```python
torch.manual_seed(0)
np.random.seed(0)
```

```python
# 配置数据集路径和参数
data_root = "/Users/xuhaohan/Desktop/neuraloperator-main/neuralop/data/datasets/data"  # 修改为你的数据路径
train_resolution = 128  # 训练数据的分辨率
test_resolutions = [128]  # 测试数据的分辨率
n_train = 100  # 训练样本数
n_tests = [20]  # 测试样本数
batch_size = 8  # 训练批次大小
test_batch_sizes = [8]  # 测试批次大小
```

```python
train_loader, test_loaders, data_processor = load_navier_stokes_pt(
    n_train=n_train,
    n_tests=n_tests,
    batch_size=batch_size,
    test_batch_sizes=test_batch_sizes,
    data_root=data_root,
    train_resolution=train_resolution,
    test_resolutions=test_resolutions,
    encode_input=True,
    encode_output=True,
)
```

```python
print(f"Train dataset size: {len(train_loader.dataset)} samples")
for res, loader in test_loaders.items():
    print(f"Test dataset at resolution={res} has {len(loader.dataset)} samples")
```

```python
fno_config = {
    "data_channels": 1,  # 输入通道数
    "out_channels": 1,   # 输出通道数
    "n_modes": [32, 32], # Fourier 模态数
    "hidden_channels": 64,  # 隐藏层通道数
    "n_layers": 6,       # FNO 层数
    "domain_padding": 0.0,  # 填充
    "fft_norm": "forward",  # FFT 归一化方式
    "norm": None,        # 归一化方式
    "skip": "linear",    # 跳跃连接
    "implementation": "reconstructed",  # 实现方式
}
model = get_model({"arch": "fno", "fno": fno_config})
```

```python
n_params = count_model_params(model)
print(f"Model parameter count: {n_params}")

# 设置设备（CPU 或 GPU）
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
# 定义优化器和学习率调度器
optimizer = AdamW(
    model.parameters(),
    lr=1e-3,  # 学习率
    weight_decay=1e-4,  # 权重衰减
)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=2, gamma=0.5)  # 学习率衰减
```

```python
# 定义损失函数
l2loss = LpLoss(d=2, p=2)  # L2 损失
h1loss = H1Loss(d=2)       # H1 损失
train_loss = h1loss        # 训练损失
eval_losses = {"h1": h1loss, "l2": l2loss}  # 评估损失
```

```python
trainer = Trainer(
    model=model,
    n_epochs=20,  # 训练轮数
    data_processor=data_processor,
    device=device,
    mixed_precision=False,  # 是否使用混合精度
    eval_interval=1,  # 评估间隔
    log_output=True,  # 是否记录输出
    verbose=True,     # 是否打印详细信息
    wandb_log=False, # 是否使用 WandB 记录
)
```

```python
print("Starting training...")
trainer.train(
    train_loader=train_loader,
    test_loaders=test_loaders,
    optimizer=optimizer,
    scheduler=scheduler,
    regularizer=False,
    training_loss=train_loss,
    eval_losses=eval_losses,
)
print("Training finished!")
```

```python

# 可视化预测结果
sample_batch = next(iter(test_loaders[test_resolutions[0]]))
x_in = sample_batch["x"].to(device)
with torch.no_grad():
    y_pred = model(x_in)
y_true = sample_batch["y"].to(device)
# 绘制结果
idx = 0  # 选择第一个样本
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(y_true[idx, 0].cpu().numpy(), cmap="viridis")
axs[0].set_title("Ground Truth")
axs[1].imshow(y_pred[idx, 0].cpu().numpy(), cmap="viridis")
axs[1].set_title("FNO Prediction")
plt.show()
```
